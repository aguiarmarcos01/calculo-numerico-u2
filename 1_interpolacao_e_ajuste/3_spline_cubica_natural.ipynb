```python
# Spline Cúbica Natural
# Autor: Marcos Vinicios Aguiar Rocha

# Dados
x = [0.0, 1.0, 2.0, 3.0]
y = [2.5, 4.5, 3.0, 6.0]

xp = 1.5

n = len(x)

# Calcula os h
h = []

for i in range(n - 1):
    h.append(x[i + 1] - x[i])

# Monta a matriz do sistema
A = []

for i in range(n):
    linha = []

    for j in range(n):
        linha.append(0)

    A.append(linha)

b = []

for i in range(n):
    b.append(0)

# Condições naturais
A[0][0] = 1
A[n - 1][n - 1] = 1

# Nós internos
for i in range(1, n - 1):

    A[i][i - 1] = h[i - 1]
    A[i][i] = 2 * (h[i - 1] + h[i])
    A[i][i + 1] = h[i]

    b[i] = 6 * (
        (y[i + 1] - y[i]) / h[i]
        - (y[i] - y[i - 1]) / h[i - 1]
    )

# Eliminação de Gauss

for k in range(n):

    pivo = A[k][k]

    for j in range(k, n):
        A[k][j] = A[k][j] / pivo

    b[k] = b[k] / pivo

    for i in range(k + 1, n):

        fator = A[i][k]

        for j in range(k, n):
            A[i][j] = A[i][j] - fator * A[k][j]

        b[i] = b[i] - fator * b[k]

# Substituição regressiva

M = []

for i in range(n):
    M.append(0)

for i in range(n - 1, -1, -1):

    soma = 0

    for j in range(i + 1, n):
        soma = soma + A[i][j] * M[j]

    M[i] = b[i] - soma

# Localiza intervalo

for i in range(n - 1):

    if x[i] <= xp <= x[i + 1]:

        hi = h[i]

        A1 = (x[i + 1] - xp) / hi
        B1 = (xp - x[i]) / hi

        C1 = ((A1 ** 3) - A1) * (hi ** 2) / 6
        D1 = ((B1 ** 3) - B1) * (hi ** 2) / 6

        yp = (
            A1 * y[i]
            + B1 * y[i + 1]
            + C1 * M[i]
            + D1 * M[i + 1]
        )

        break

print("Spline Cúbica Natural")
print("Posição em t =", xp)
print("y =", yp)
```
