```python
# Interpolação por Spline Linear
# Autor: Marcos Vinicios Aguiar Rocha

# Dados
x = [0.0, 1.0, 2.0, 3.0]
y = [2.5, 4.5, 3.0, 6.0]

# Ponto a interpolar
xp = 1.5

# Procura o intervalo correto
for i in range(len(x) - 1):

    if x[i] <= xp <= x[i + 1]:

        # coeficiente angular
        m = (y[i + 1] - y[i]) / (x[i + 1] - x[i])

        # interpolação linear
        yp = y[i] + m * (xp - x[i])

        break

print("Spline Linear")
print("Posição em t =", xp)
print("y =", yp)
```
