```python
# Interpolação de Gregory-Newton Progressiva
# Autor: Marcos Vinicios Aguiar Rocha

# Dados
x = [10, 20, 30, 40]
y = [45.0, 52.0, 60.0, 71.0]

# Valor a ser interpolado
xp = 25

n = len(x)
h = x[1] - x[0]

# Construção da tabela de diferenças finitas
tabela = []

for i in range(n):
    tabela.append([y[i]])

for j in range(1, n):
    for i in range(n - j):
        diferenca = tabela[i + 1][j - 1] - tabela[i][j - 1]
        tabela[i].append(diferenca)

# Cálculo de u
u = (xp - x[0]) / h

# Inicialização do polinômio
resultado = tabela[0][0]

produto = 1
fatorial = 1

for i in range(1, n):

    produto = produto * (u - (i - 1))
    fatorial = fatorial * i

    resultado = resultado + (produto / fatorial) * tabela[0][i]

print("Temperatura estimada =", resultado, "°C")
```
