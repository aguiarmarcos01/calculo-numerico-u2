```python
# Ajuste Linear pelo Método dos Mínimos Quadrados (MMQ)
# Autor: Marcos Vinicios Aguiar Rocha

# Dados
x = [8, 9, 10, 11, 12]
y = [2.1, 2.8, 3.1, 4.0, 4.8]

n = len(x)

# Inicialização das somas
soma_x = 0
soma_y = 0
soma_x2 = 0
soma_xy = 0

# Cálculo dos somatórios
for i in range(n):
    soma_x = soma_x + x[i]
    soma_y = soma_y + y[i]
    soma_x2 = soma_x2 + x[i] * x[i]
    soma_xy = soma_xy + x[i] * y[i]

# Coeficiente angular
a = (n * soma_xy - soma_x * soma_y) / (n * soma_x2 - soma_x * soma_x)

# Coeficiente linear
b = (soma_y - a * soma_x) / n

print("Equação da reta:")
print("P(x) =", a, "* x +", b)

# Previsão para x = 13
xp = 13
yp = a * xp + b

print()
print("Previsão para x =", xp)
print("P(", xp, ") =", yp)
```
