```python
# Integração Numérica
# Regra 3/8 de Simpson

# Dados
t = [0, 2, 4, 6]
v = [10, 15, 12, 8]

# Passo
h = t[1] - t[0]

# Aplicação da fórmula 3/8 de Simpson
integral = (3 * h / 8) * (
    v[0]
    + 3 * v[1]
    + 3 * v[2]
    + v[3]
)

print("Total de dados transferidos:")
print(integral, "MB")
```
