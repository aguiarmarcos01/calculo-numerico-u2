```python
# Integração Numérica
# Trapézios Repetida e Simpson 1/3 Repetida

t = [0.0, 0.5, 1.0, 1.5, 2.0]
v = [0, 40, 65, 80, 90]

h = t[1] - t[0]

# -------------------------
# Trapézios Repetida
# -------------------------

soma_interna = 0

for i in range(1, len(v) - 1):
    soma_interna = soma_interna + v[i]

trap = (h / 2) * (v[0] + 2 * soma_interna + v[-1])

# -------------------------
# Simpson 1/3 Repetida
# -------------------------

soma_impares = 0
soma_pares = 0

for i in range(1, len(v) - 1):

    if i % 2 == 1:
        soma_impares = soma_impares + v[i]
    else:
        soma_pares = soma_pares + v[i]

simpson = (h / 3) * (
    v[0]
    + 4 * soma_impares
    + 2 * soma_pares
    + v[-1]
)

print("Trapézios Repetida =", trap, "km")
print("Simpson 1/3 Repetida =", simpson, "km")
```
