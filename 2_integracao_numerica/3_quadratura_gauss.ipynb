```python
# Quadratura de Gauss-Legendre
# Implementação manual para n = 2 e n = 3
# Autor: Marcos Vinicios Aguiar Rocha

def quadratura_gauss(funcao, a, b, n_pontos):

    if n_pontos == 2:

        pontos = [
            -1 / (3 ** 0.5),
             1 / (3 ** 0.5)
        ]

        pesos = [1, 1]

    elif n_pontos == 3:

        pontos = [
            -(3 / 5) ** 0.5,
             0,
             (3 / 5) ** 0.5
        ]

        pesos = [
            5 / 9,
            8 / 9,
            5 / 9
        ]

    else:
        print("Somente n = 2 ou n = 3")
        return None

    soma = 0

    for i in range(len(pontos)):

        # transformação para [a,b]
        x = ((b - a) / 2) * pontos[i] + ((a + b) / 2)

        soma = soma + pesos[i] * funcao(x)

    integral = ((b - a) / 2) * soma

    return integral


# Função do problema
def f(x):
    return 5 * (x ** 3) + (x ** 2) - 12 * x + 4


# Dados
a = -1
b = 1

resultado = quadratura_gauss(f, a, b, 2)

print("Integral =", resultado)
```
